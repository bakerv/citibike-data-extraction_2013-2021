```python
import wget
import zipfile
import os
import shutil
import pandas as pd
```

```python
file_name = '201306-citibike-tripdata'
save_directory = '../../../Data_Sets/citibike/'

def citibike_extraction(save_directory, file_name):
    zip_file_path = save_directory + file_name + '.zip'
    csv_path =  save_directory + file_name + '.csv'
    
    # download zip files
    def zip_download(zip_file_path, file_name):
        url = 'https://s3.amazonaws.com/tripdata/' + file_name + '.zip'
        wget.download(url, zip_file_path)
     
    # extract the windows version of the csv file from the zip directory
    def extract_csv(zip_file_path, save_directory):
        zip_contents = zipfile.ZipFile(zip_file_path,'r')
        zip_contents.extractall(save_directory)
        
    # delete the original zip file to reclaim disk space
    def clean_up(zip_file_path, save_directory):
        os.remove(zip_file_path)
        shutil.rmtree(save_directory + '__MACOSX')
        
     # rename columns and entries to match 2021 formating 
    def clean_data1(csv_path):
        # rename columns to match 2021 headers

        csv_df = pd.read_csv(csv_path)
        clean_df = csv_df.copy().rename(columns={
            'bike_id':'ride_id',
            'starttime':'started_at',
            'stoptime':'ended_at',
            'start station name':'start_station_name',
            'start station id':'start_staton_id',
            'end station name':'end_station_name',
            'end station id':'end_station_id',
            'start station latitude':'start_lat',
            'start station longitude':'start_lng',
            'end station latitude':'end_lat',
            'end station longitude':'end_lng',
            'usertype':'member_casual',
            'birth year':'Birth Year',
            'gender':'Gender',
            'trip duration':'Trip Duration'
        })

        #replace membership designations with terms used in the 2021 dataset
        clean_df['member_casual'] = clean_df['member_casual'].replace({
            'Subscriber':'member',
            'Customer':'casual'
        })
        
        #overwrite the orginal csv, with the clean_csv
        clean_df.to_csv(csv_path)
        
        
    zip_download(zip_file_path, file_name)
    extract_csv(zip_file_path, save_directory)
    clean_up(zip_file_path, save_directory)
    clean_data1(csv_path)
```
